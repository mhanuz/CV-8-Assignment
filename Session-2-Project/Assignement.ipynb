```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
main_img = Image.open('Cat.jpg')
```

```python
plt.imshow(main_img)
plt.show()
```

```python
# Cropping 
crop_box = (50, 50, 200, 200)
cropped_img = main_img.crop(crop_box)
```

```python
plt.imshow(cropped_img)
plt.show()
```

```python
# Resizing
resized_img = main_img.resize((900, 600))
```

```python
plt.imshow(resized_img)
plt.show()
```

```python
#rescaling
scale_factor = 0.5  # 50% of original size
new_size = (int(main_img.width * scale_factor), int(main_img.height * scale_factor))
rescaled_img = main_img.resize(new_size)
```

```python
plt.imshow(rescaled_img)
plt.show()
```

```python
# Horizontal Flip (Left to Right)
flipped_horizontally = main_img.transpose(Image.FLIP_LEFT_RIGHT)

```

```python
plt.imshow(flipped_horizontally)
plt.show()
```

```python
# Vertical Flip (Top to Bottom)
flipped_vertically = main_img.transpose(Image.FLIP_TOP_BOTTOM)
```

```python
plt.imshow(flipped_vertically)
plt.show()
```

```python
# Rotate 90 degrees counter-clockwise
rotated_img = main_img.rotate(90, expand=True)
```

```python
plt.imshow(rotated_img)
plt.show()
```
